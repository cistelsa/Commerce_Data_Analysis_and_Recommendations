# **Filtrado de Datasets pre ETL**

```python
# Ejemplo de Lectura del dataset con Pandas
import pandas as pd
# Oficial Lectura del dataset con Spark
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, DoubleType, TimestampType
```

```python
df_review = pd.read_json('/lakehouse/default/Files/data/original/Yelp/review.json', lines=True)
```

```python

# Define manualmente el esquema del JSON
json_schema = StructType([
    StructField("review_id", StringType(), True),
    StructField("user_id", StringType(), True),
    StructField("business_id", StringType(), True),
    StructField("stars", IntegerType(), True),
    StructField("useful", IntegerType(), True),
    StructField("funny", IntegerType(), True),
    StructField("cool", IntegerType(), True),
    StructField("text", StringType(), True),
    StructField("date", TimestampType(), True)
])
```

```python
# Lee el archivo JSON por líneas y aplica el esquema
df_review = spark.read.schema(json_schema).json("Files/data/original/Yelp/review.json")
```

```python
df_review.count()
```

### Dataset `business.csv`

```python
# Oficial Lectura del dataset con Spark
df_business = spark.read.format("csv").option("header","true").load("Files/data/original/Yelp/business.csv")

display(df_business)
```

```python
df_review = df_review.join(df_business, on="business_id", how="inner")
```

```python
df_review.count()
```

## Lectura dataset tip.json

```python
# Define manualmente el esquema del JSON
json_schema_tip = StructType([
    StructField("user_id", StringType(), True),
    StructField("business_id", StringType(), True),
    StructField("text", StringType(), True),
    StructField("date", TimestampType(), True),
    StructField("compliment_count", IntegerType(), True),
])
```

```python
# Lee el archivo JSON por líneas y aplica el esquema
df_tip = spark.read.schema(json_schema_tip).json("Files/data/original/Yelp/tip.json")
```

```python
display(df_tip.limit(1))
```

```python
# Inner Join para filtrar los datos
df_tip = df_tip.join(df_business.select("business_id"), on="business_id", how="inner")
```

```python
df_tip.count()
```

```python
# Convertir a DF Pandas
df_tip_p = df_tip.toPandas()
```

```python
# Guardar Datframe en archivo CSV
df_tip_p.to_csv("/lakehouse/default/Files/data/original/Yelp/tip_f.csv", index=False)
```

## Lectura checkin.json

```python
# Define manualmente el esquema del JSON
json_schema_checkin = StructType([
    StructField("business_id", StringType(), True),
    StructField("date", StringType(), True)
])
```

```python
# Lee el archivo JSON por líneas y aplica el esquema
df_checkin = spark.read.schema(json_schema_checkin).json("Files/data/original/Yelp/checkin.json")
```

```python
# Inner Join para filtrar los datos
df_checkin = df_checkin.join(df_business.select("business_id"), on="business_id", how="inner")
```

```python
df_checkin.count()
```

```python
# Convertir a DF Pandas
df_checkin_p = df_checkin.toPandas()
```

```python
# Guardar Datframe en archivo CSV
df_checkin_p.to_csv("/lakehouse/default/Files/data/original/Yelp/checkin_f.csv", index=False)
```
